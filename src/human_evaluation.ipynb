```python
import pandas as pd
import numpy as np
```

```python
chatgpt = pd.read_json("../data/annotate/chatgpt.json", orient="records")
our_model = pd.read_json("../data/annotate/our_model.json", orient="records")
alpaca = pd.read_json("../data/annotate/alpaca.json", orient="records")
```

```python
label2df = {0: chatgpt, 1: our_model, 2: alpaca}
label2name = {0: "chatgpt", 1: "our_model", 2: "alpaca"}
```

```python
inputs = []
```

```python
counter = {"chatgpt": 0, "our_model": 0, "alpaca": 0}
```

```python
for i in range(len(chatgpt)):
    val = np.random.choice([0,1,2])

    df1, name1 = label2df[val], label2name[val]
    df2, name2 = label2df[(val+1)%3], label2name[(val+1)%3]
    df3, name3 = label2df[(val+2)%3], label2name[(val+2)%3]


    a1 = df1.iloc[i]["model_answer"]
    a2 = df2.iloc[i]["model_answer"]
    a3 = df3.iloc[i]["model_answer"]

    print("============================")
    print("Which generation do you prefer:")
    print("0: ", a1)
    print()
    print("1: ", a2)
    print()
    print("2: ", a3)
    user_preference = int(input())
    
    if user_preference == 0:
        counter[name1] += 1
    elif user_preference == 1:
        counter[name2] += 1
    else:
        counter[name2] += 1
    
```
